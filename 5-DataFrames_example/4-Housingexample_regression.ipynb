```julia
using DataFrames
df = readtable("houses.csv")
df = df[df[:sq__ft].>0,:]
X = df[[:beds,:baths,:sq__ft]]
y = df[:price]
m = length(y)
@show [X y]
```

```julia
mu_vals = zeros(size(X,2))
std_vals = zeros(size(X,2))
for i = 1:size(X,2)
    mu_vals[i] = mean(X[:,i])
    std_vals[i] = std(X[:,i])
    X[:,i] = (X[:,i] - mu_vals[i])/std_vals[i]
end
```

```julia
intercept = DataFrame(intercept=ones(Int64,m))
P = [intercept  X]
```

```julia
α = 0.015
num_iters = 100
n = size(P,2)
θ = zeros(n)
history = zeros(num_iters, 1)
```

```julia
X = Array(P)
```

```julia
for iter = 1:num_iters
    δ = 1/m*(X'*X*θ-X'*y)
    θ = θ - α.*δ
    v = X*θ
    j = 1/(2*m)*sum((v-y).^2)
    history[iter] = j
end
```

```julia
d = [1.0,2,1,836]
d[2:end] =  d[2:end] - mu_vals 
d[2:end] = d[2:end]./std_vals
price = dot(d,θ)
```
