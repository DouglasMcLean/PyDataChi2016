```julia
using RDatasets
using Clustering
using DataFrames
using MultivariateStats
using Plots
pyplot(legend=false)
```

```julia
iris = dataset("datasets", "iris")
```

```julia
X = Array(iris[:, 1:4])
X = X'
result = kmeans(X,3)

tf = result.assignments.==1
x = X[3,tf]
y = X[4,tf]
scatter(x,y, color = "red")

tf = result.assignments.==2
x = X[3,tf]
y = X[4,tf]
scatter!(x,y, color = "green")

tf = result.assignments.==3
x = X[3,tf]
y = X[4,tf]
scatter!(x,y, color = "blue")
```

```julia
pc = fit(PCA, X; maxoutdim = 2)
pca_result = transform(pc, X)
x = pca_result[1,:]';
y = pca_result[2,:]';
scatter(x,y,color = "green")
```
